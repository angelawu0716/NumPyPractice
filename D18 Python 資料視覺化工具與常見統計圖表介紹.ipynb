```python
# ...
import matplotlib.pyplot as plt
import numpy as np

#  ... sin 0~180
# x,ylist....
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)

# 
plt.plot(x,y)

#  
#  ""
plt.show() 

# sin 
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()

# cos 

y_cos = np.cos(x)

plt.plot(x, y_cos)

# :
#  pyplot.show() :

plt.savefig("filename.png",dpi=300,format="png")
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)


plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

area = np.random.rand(n) * 1000

plt.scatter(X, Y, c=area, cmap='BrBG')
plt.show()
```
